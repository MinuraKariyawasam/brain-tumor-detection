```python
import cv2
from matplotlib import pyplot as plt

img = cv2.imread('test_images/0 (239).jpg',0)
print()

plt.imshow(img, cmap='gray')
```

```python
print(img)
```

```python
height, width = img.shape

imgFlat = img.reshape(height*width,1)

print(imgFlat.shape)
```

```python
from sklearn.cluster import KMeans

model = KMeans(n_clusters=5)

model.fit(imgFlat)
```

```python
lables = model.labels_
lable2D = lables.reshape(height,width)
```

```python
plt.imshow(lable2D)
plt.colorbar()
plt.show()
```

```python
# save the model
import joblib
joblib.dump(model,'brain_tumor_detection_model.sav')
```
